```python
import pandas as pd
df = pd.DataFrame()

for i in range(1,3):
  url = "https://event.moc.gov.tw/sp.asp?xdurl=ccEvent2016/eventSearchList.asp&ev_start=&ev_end=&dateLimit=&ev_city=&ev_char1=all&ev_char2=&ev_char3=&stitle=&ev_series_id=&ev_place=&ev_location=&ev_format=all&ev_shower=&dept_name=&ev_xx=&pageSize=10&groupingField=ev_city_g&groupingValue=%E8%87%BA%E5%8C%97%E5%B8%82&ctNode=676&mp=1&currentPage="+str(i)
  df = pd.concat([df,pd.read_html(url)[0]])
  print("第{page}頁完成~".format(page=i+1))

df = df[['序號', '日期', '活動名稱', '縣市', '地點', '活動類別']]
df.set_index('序號', inplace=True)
```

```python
df.to_csv('test.csv')
```
